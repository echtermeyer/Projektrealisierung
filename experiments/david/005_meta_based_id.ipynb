```python
def get_action_occurence_per_id(df, id_col):
    action_count_series = df.drop_duplicates(subset=[id_col, 'action_name'])['action_name'].value_counts().reset_index()
    action_count_df = pd.DataFrame(action_count_series)
    action_count_df["percent"] = action_count_df["count"] / df[id_col].nunique()
    return action_count_df
```

```python
def get_ids_without_certain_action(df, id_col, action_name):
    grouped = df.groupby(id_col)
    grouped_count = grouped["action_name"].value_counts().reset_index()
    grouped_count_filtered = grouped_count.query(f"action_name == '{action_name}'")
    ids_with_action = set(grouped_count_filtered[id_col].unique())
    ids_without_action = set(df[id_col].unique()) - ids_with_action
    return ids_without_action
```

Es kommen regelmäßig mehrere Nachrichten auf einmal rein (exakt selber timestamp). Nachrichten innerhalb eines solchen batches haben dieselbe header_id. 
Die ersten nachrichten des batches haben category received und anderen haben category saved. Je nach action_name der reveived message wird ein spezifischer action_name als saved message gespeichert.

Die letzte received message gehört zu der ersten saved message, die vorletzte received message zur zweiten saved message usw. 
Ausnahmen: ResetLoadingListRecordsAction, UpdateCrewDataAction und EstimateStorePaxDataAction lösen on receive keine saved message aus, CreateZFWMessageAction löst unmittelbar eine send message aus, und später dann eine saved message (nach der oben beschriebenen Reihenfolge)
